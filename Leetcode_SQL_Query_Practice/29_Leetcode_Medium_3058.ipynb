

## Problem: **Friends With No Mutual Friends** (LeetCode 3058)

**Description**

We have a table **Friends** with columns:

| Column Name | Type |
| ----------- | ---- |
| user_id1    | int  |
| user_id2    | int  |

Each row `(user_id1, user_id2)` indicates that `user_id1` and `user_id2` are friends (i.e. an undirected friendship relation). The pair `(user_id1, user_id2)` is the primary key, so each friendship appears exactly once.

Write a query (or algorithm) to find all pairs of users who:

1. Are **friends with each other**, and
2. Have **no mutual friends** (i.e. there is no third user who is a friend of both).

Return the result ordered by `user_id1`, `user_id2` in ascending order.

---

### Explanation and Key Points

* “Mutual friend” means a third user `x` such that `x` is a friend of `user_id1` *and* `x` is a friend of `user_id2`.
* Because the friendship is undirected, from table you should consider that `(u, v)` implies both directions: `u` ↔ `v`.
* The output should list only those friend-pairs that **do not** share any other friend in common.

---

### Example from problem

**Input:**

Friends table:

| user_id1 | user_id2 |
| -------- | -------- |
| 1        | 2        |
| 2        | 3        |
| 2        | 4        |
| 1        | 5        |
| 6        | 7        |
| 3        | 4        |
| 2        | 5        |
| 8        | 9        |

**Output:**

| user_id1 | user_id2 |
| -------- | -------- |
| 6        | 7        |
| 8        | 9        |

**Explanation:**

* (1,2) are friends, but they share a mutual friend 5 → exclude.
* (2,3) share mutual friend 4 → exclude.
* (2,4) share mutual friend 3 → exclude.
* (1,5) share mutual friend 2 → exclude.
* (3,4) share mutual friend 2 → exclude.
* (2,5) share mutual friend 1 → exclude.
* (6,7) have no mutual friend → include.
* (8,9) have no mutual friend → include.
