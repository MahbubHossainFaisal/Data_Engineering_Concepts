# 3278 — Find Candidates for Data Scientist Position II

**Source:** LeetCode problem (mirrored). ([Leetcode][1])

---

## Problem

You are given two tables, `Candidates` and `Projects`. For each project, find the best candidate who satisfies **all required skills** for that project and yields the highest score computed as described below. Return one row per project (the top candidate), ordered by `project_id` ascending. If there is a tie on score, choose the candidate with the lower `candidate_id`. If no candidate has all required skills for a project, do not return that project.

### Tables

**Candidates**

```
+--------------+---------+
| Column Name  | Type    |
+--------------+---------+
| candidate_id | int     |
| skill        | varchar |
| proficiency  | int     |
+--------------+---------+
```

Primary key: `(candidate_id, skill)` — each row describes one candidate's proficiency for a particular skill.

**Projects**

```
+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| project_id  | int     |
| skill       | varchar |
| importance  | int     |
+-------------+---------+
```

Primary key: `(project_id, skill)` — each row describes one required skill and its importance (1–5) for a given project.

---

## Scoring rules (for a candidate on a project)

* Start with `100` points.
* For each required skill:

  * If `proficiency > importance` → **add 10** points.
  * If `proficiency < importance` → **subtract 5** points.
  * If `proficiency = importance` → **add 0** points.

Only consider candidates who have **all** the skills required by the project.

---

## Output

Return a table with the columns:

```
+------------+--------------+-------+
| project_id | candidate_id | score |
+------------+--------------+-------+
```

One row per project (the chosen top candidate), ordered by `project_id` ascending.

---

## Example

**Input**

`Candidates`:

```
candidate_id | skill       | proficiency
---------------------------------------
101          | Python      | 5
101          | Tableau     | 3
101          | PostgreSQL  | 4
101          | TensorFlow  | 2
102          | Python      | 4
102          | Tableau     | 5
102          | PostgreSQL  | 4
102          | R           | 4
103          | Python      | 3
103          | Tableau     | 5
103          | PostgreSQL  | 5
103          | Spark       | 4
```

`Projects` (required skills with importance):

```
project_id | skill       | importance
------------------------------------
501        | Python      | 4
501        | Tableau     | 3
501        | PostgreSQL  | 4
501        | TensorFlow  | 2

502        | Python      | 3
502        | Tableau     | 5
502        | PostgreSQL  | 4
```

**Output**

```
project_id | candidate_id | score
--------------------------------
501        | 101          | 105
502        | 102          | 130
```

**Explanation**

* For project `501`, candidate `101` has all required skills and the top score (ties broken by lower `candidate_id`).
* For project `502`, candidate `102` has the highest score.
