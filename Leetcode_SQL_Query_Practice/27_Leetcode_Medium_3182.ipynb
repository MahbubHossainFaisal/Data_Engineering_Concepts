# 3182 — Find Top Scoring Students

**Source:** LeetCode problem page. ([Leetcode][1])

---

## Problem (cleanly formatted)

Given three tables — `students`, `courses`, and `enrollments` — return the `student_id` of every student who received an **'A'** grade in **every course required by their major**. Order the output by `student_id` ascending.

### Tables

**students**

```
student_id  | int
name        | varchar
major       | varchar
```

**courses**

```
course_id   | int
name        | varchar
credits     | int
major       | varchar
```

* The `courses` table lists required courses per major (i.e., each course belongs to a major).

**enrollments**

```
student_id  | int
course_id   | int
semester    | varchar
grade       | varchar
```

* Each row records one student’s grade in one course during a semester.

---

## Example

**enrollments**

```
student_id | course_id | semester   | grade
-----------+-----------+------------+------
1          | 101       | Fall 2023  | A
1          | 102       | Fall 2023  | A
2          | 101       | Fall 2023  | B
2          | 102       | Fall 2023  | A
3          | 103       | Fall 2023  | A
3          | 104       | Fall 2023  | A
4          | 103       | Fall 2023  | A
4          | 104       | Fall 2023  | B
```

(Implicit sample `students` and `courses` for the example)

```
students:
1 | Alice   | Computer Science
2 | Bob     | Computer Science
3 | Charlie | Mathematics
4 | David   | Mathematics

courses (per major):
101 | Algorithms       | ... | Computer Science
102 | Data Structures  | ... | Computer Science
103 | Calculus         | ... | Mathematics
104 | Linear Algebra   | ... | Mathematics
```

**Expected output**

```
student_id
----------
1
3
```

---

## Explanation (example)

* Alice (student_id 1) is a Computer Science major and took both required CS courses (101, 102) and received an 'A' in both → included.
* Bob (student_id 2) did not receive an 'A' in all required CS courses (one grade is 'B') → excluded.
* Charlie (student_id 3) is a Math major and received 'A' in both required Math courses → included.
* David (student_id 4) did not receive an 'A' in all required Math courses → excluded.

**Note:** Output must be ordered by `student_id` ascending. ([Leetcode][1])

````python
WITH required AS (
  SELECT s.student_id, s.name, s.major, c.course_id
  FROM students s
  JOIN courses c
    ON c.major = s.major            
)
SELECT r.student_id
FROM required r
LEFT JOIN enrollments e
  ON e.student_id = r.student_id
 AND e.course_id  = r.course_id
 AND e.grade = 'A'                   
GROUP BY r.student_id
HAVING COUNT(DISTINCT r.course_id) = COUNT(DISTINCT e.course_id)
ORDER BY r.student_id;
````
